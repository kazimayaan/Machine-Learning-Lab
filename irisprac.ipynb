```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def KNN(data):
  print("enter the petal sepal lololol")
  a,b,c,d = input().split()
  a = float(a)
  b = float(b)
  c = float(c)
  d = float(d)
  for line in data:
    x = float(line[0])
    y = float(line[1])
    z = float(line[2])
    w = float(line[3])
    dist = np.sqrt((x-a)**2+(y-b)**2+(z-c)**2+(w-d)**2)
    line.append(dist)
  data.sort(key = lambda i:i[5])
  display(data)
  setosa = 0
  vir = 0
  ver = 0
  k = input('enyer k')
  k = int(k)
  for i in range(k):
    if data[i][4]=='Iris-setosa':
      setosa+=1
    elif data[i][4]=='Iris-versicolor':
      ver+=1
    elif data[i][4]=='Iris-virginica':
      vir+=1

  lol = [setosa,ver,vir]
  label = ['setosa','versicolor','virginica']
  plt.bar(label,lol)
  plt.show()

```

```python
file = "/content/iris.csv"
temp = pd.read_csv(file)
display(temp)
fd = csv.reader(open(file))

data=[]
for line in fd:
  data.append(line)
Data = data[1:]
display(Data)
KNN(Data)
```
