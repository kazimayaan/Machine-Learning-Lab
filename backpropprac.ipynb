```python
import numpy as np
```

```python
IPN = 2;
HLN = 4
OPN = 2
iteration = 6000


input = np.random.randint(1,5,IPN)
output = np.array([1.0,0.0])
hiddenlayer = np.random.rand(1,HLN)

hidden_bias = np.random.rand(1,HLN)
output_bias = np.random.rand(1,OPN)

hidden_weight = np.random.rand(IPN,HLN)
output_weight = np.random.rand(HLN,OPN)

def sigmoid(layer):
  return 1/(1+ np.exp(-layer))

def gradient(layer):
  return layer*(1-layer)

for i in range(iteration):
  hiddenlayer = np.dot(input,hidden_weight)
  hiddenlayer = sigmoid(hiddenlayer+hidden_bias)

  outputlayer = np.dot(hiddenlayer,output_weight)
  outputlayer = sigmoid(outputlayer+output_bias)

  error = output - outputlayer
  g_outputlayer= gradient(outputlayer)
  g_hiddenlayer = gradient(hiddenlayer)
  error_terms_output = g_outputlayer*error
  error_terms_hidden = g_hiddenlayer*np.dot(error_terms_output,output_weight.T)

  gradient_hidden_weights = np.dot(input.reshape(IPN,1),error_terms_hidden.reshape(1,HLN))
  gradient_output_weights = np.dot(hiddenlayer.reshape(HLN,1),error_terms_output.reshape(1,OPN))

  hidden_weight = hidden_weight+ 0.5*gradient_hidden_weights
  output_weight = output_weight+ 0.5*gradient_output_weights

  print("iteration",i)
  print("error terms is ",error)
  print("output is ",outputlayer)

  print()

  
```
