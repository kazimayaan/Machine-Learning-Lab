```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
def KNN(DATA):
  print("Enter your ,heightchest size,eight to buy the right size")
  a,b,c=input().split()
  a=int(a)
  b=int(b)
  c=int(c)
  for line in DATA:
    x= int(line[0])
    y= int(line[1])
    z= int(line[2])
    dist= np.sqrt((x-a)**2+(y-b)**2+(z-c)**2)
    line.append(dist)

  # for line in DATA:
  #   print(line)
  print('how many nearesr surbvey key?>')
  k = int(input())
  DATA.sort(key= lambda i:i[4])
  scount=0
  mcount=0
  lcount=0
  xlcount=0

  for j in range(k):
    if DATA[j][3]=='S':
      scount+=1
    if DATA[j][3]=='M':
      mcount+=1
    if DATA[j][3]=='L':
      lcount+=1
    if DATA[j][3]=='XL':
      xlcount+=1

  x1=["S",'M','L','XL']
  y1=[scount,mcount,lcount,xlcount]
  plt.bar(x1,y1)
  plt.show()
  if(scount>lcount and scount>mcount and scount>xlcount):
    print("--------------GO FOR S SIZE--------------")
  elif(lcount>scount and lcount>mcount and lcount>xlcount):
    print("--------------GO FOR L SIZE--------------")
  elif(mcount>lcount and mcount>scount and mcount>xlcount):
    print("--------------GO FOR M SIZE--------------")
  else:
    print("--------------GO FOR XL SIZE--------------")
```
