```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/tennis.csv')
display(df)
d = {'Sunny':0,'Rainy':1,'Overcast':2}
df['Outlook'] = df['Outlook'].map(d)
d = {'Hot':0, 'Mild':1, 'Cool':2}
df['Temperature'] = df['Temperature'].map(d)
d = {'High':0,'Normal':1}
df['Humidity'] = df['Humidity'].map(d)
d = {'No':1,'Yes':0}
df['PlayTennis']= df['PlayTennis'].map(d)
display(df)
features = ['Outlook','Temperature','Humidity','Windy']
x = df[features]
y = df['PlayTennis']

print(x)
print(y)

dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)
tree.plot_tree(dtree,feature_names= features)
```
