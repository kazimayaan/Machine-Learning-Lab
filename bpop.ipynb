```python
import numpy as np
```

```python
ipn = 2
opn = 2
hln = 4
it = 6000

input = np.random.randint(1,5,ipn)
output = np.array([1.0,0.0])
hiddenlayer = np.random.rand(1,hln)

hiddenbias = np.random.rand(1,hln)
outputbias = np.random.rand(1,opn)

hiddenweight = np.random.rand(ipn,hln)
outputweight = np.random.rand(hln,opn)

def sigmoid(layer):
  return 1/(1+np.exp(-layer))

def gradient(layer):
  return layer*(1-layer)

for i in range(it):
  hiddenlayer = np.dot(input,hiddenweight)
  hiddenlayer = sigmoid(hiddenlayer + hiddenbias)

  outputlayer = np.dot(hiddenlayer,outputweight)
  outputlayer = sigmoid(outputlayer+outputbias)

  error = output - outputlayer

  gop = gradient(outputlayer)
  goh = gradient(hiddenlayer)

  errop = error * gop
  errhl = goh * np.dot(errop,outputweight.T)

  grhl = np.dot(input.reshape(ipn,1),errhl.reshape(1,hln))
  grop = np.dot(hiddenlayer.reshape(hln,1),errop.reshape(1,opn))

  hiddenweight = hiddenweight + 0.5*grhl
  outputweight = outputweight +0.5*grop

  if i < 50 or i > 5500:
    print("----")
    print("error:",error)
    print("###output###",outputlayer)
```
