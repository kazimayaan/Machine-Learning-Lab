```python
import numpy as np
```

```python
ipn = 2
opn = 2
hln = 4
it = 6000

input = np.random.randint(1,5,ipn)
output = np.array([1.0,0.0])
hiddenlayer = np.random.rand(1,hln)

hiddenbias = np.random.rand(1,hln)
outputbias = np.random.rand(1,opn)

hidden_weight = np.random.rand(ipn,hln)
output_weight = np.random.rand(hln,opn)

def sigmoid(layer):
  return 1/(1+ np.exp(-layer))

def gradient(layer):
  return layer*(1-layer)

for i in range(it):
  hiddenlayer = np.dot(input,hidden_weight)
  hiddenlayer = sigmoid(hiddenlayer+hiddenbias)

  outputlayer = np.dot(hiddenlayer,output_weight)
  outputlayer = sigmoid(outputlayer+outputbias)

  error = output-outputlayer

  gop = gradient(outputlayer)
  goh = gradient(hiddenlayer)

  errorop = error*gop
  errorhl = goh * np.dot(errorop,output_weight.T)

  grdh = np.dot(input.reshape(ipn,1),errorhl.reshape(1,hln))
  grdo = np.dot(hiddenlayer.reshape(hln,1),errorop.reshape(1,opn))

  hidden_weight = hidden_weight + 0.5*grdh
  output_weight = output_weight + 0.5*grdo

  if i<50 or i>5500:
    print("--------")
    print(error)
    print(outputlayer)




```
