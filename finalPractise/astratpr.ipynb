```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge('A','B',weight = 5)
G.add_edge('A','c',weight = 15)
G.add_edge('A','K',weight = 125)
G.add_edge('A','V',weight = 5)
G.add_edge('V','F',weight = 5)
G.add_edge('F','M',weight = 5)
G.add_edge('M','K',weight = 5)

path = nx.astar_path(G,'A','K',weight='weight')
print(path)
nx.draw(G,with_labels='yes')
```
