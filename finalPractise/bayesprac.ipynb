```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
def Bayes(data,x,col,yescount,nocount):
  xyes = 0
  xno = 0
  for line in data:
    if line[col]==x:
      if line[-1]=='yes':
        xyes+=1
      else:
        xno+=1

  pxyes = xyes/yescount
  pxno = xno/nocount
  return pxyes,pxno
```

```python
file ='/content/car (2).csv'
temp = pd.read_csv(file)
display(temp)
Data = []
fd = csv.reader(open(file))
for line in fd:
  Data.append(line)

data = Data[1:]
display(data)

yescount = 0
nocount = 0

for line in data:
  if line[-1]=='yes':
    yescount+=1
  else:
    nocount+=1
n = len(data)
pyes = yescount/n
pno = nocount/n

print("enter the color type and origiun")
x,y,z = input().split()
xyes,xno = Bayes(data,x,0,yescount,nocount)
yyes,yno = Bayes(data,y,1,yescount,nocount)
zyes,zno = Bayes(data,z,2,yescount,nocount)

resyes = pyes*xyes*yyes*zyes
resno = pno*xno*yno*zno

peryes = (resyes/(resyes+resno))*100
perno = (resno/(resyes+resno))*100

pex = [peryes,perno]
label = ['Cat stolne','Not stolem']

plt.pie(pex,labels = label)
plt.show()
```
