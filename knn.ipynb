```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def knn(data):
  a,b,c = input("enyer tyhje weigh height and ches stize").split()
  a = int(a)
  b = int(b)
  c = int(c)

  for line in data:
    x = int(line[0])
    y = int(line[1])
    z = int(line[2])
    dist = np.sqrt((x-a)**2+(y-b)**2+(z-c)**2)
    line.append(dist)
  data.sort(key = lambda i:i[4])
  display(data)
  sc = 0
  mc = 0
  lc = 0
  xc = 0
  k = input("enter k ")
  k = int(k)

  for i in range (k):
    if data[i][3]=='S':
      sc+=1
    elif data[i][3]=='M':
      mc+=1
    elif data[i][3]=='L':
      lc+=1
    else:
      xc+=1

  lol = [sc,mc,lc,xc]
  ylan = ['S','M','L','XL']
  plt.bar(ylan,lol)
  plt.show()
```

```python
file = "/content/T-shirt1.csv"
temp = pd.read_csv(file)
display(temp)
fd = csv.reader(open(file))
data = []
for line in fd:
  data.append(line)

knn(data[1:])
```
