```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def KNN(data):
  print("enter weight height and chest size")
  a,b,c = input().split()
  a = int(a)
  b= int(b)
  c = int(c)
  for line in data:
    x = int(line[0])
    y = int(line[1])
    z = int(line[2])
    dist = np.sqrt((x-a)**2+(y-b)**2+(z-c)**2)
    line.append(dist)
  data.sort(key = lambda i:i[4])
  display(data)
  sc = 0
  mc = 0
  lc = 0
  xc = 0
  print("entyer k ")
  k = int(input())
  for i in range(k):
    if data[i][-2]=='S':
      sc+=1
    if data[i][-2]=='M':
      mc+=1
    if data[i][-2]=='L':
      lc+=1
    if data[i][-2]=='XL':
      xc+=1

  lol = [sc,mc,lc,xc]
  label = ['S','M','L','XL']

  plt.bar(label,lol)
  plt.xlabel("sizes")
  plt.ylabel("closest matches")
  plt.show()
```

```python
file = "/content/T-shirt1.csv"
temp = pd.read_csv(file)
display(temp)
fd = csv.reader(open(file))
data = []
for line in fd:
  data.append(line)

Data = data[1:]
KNN(Data)
```
