```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def KNN(data):
  print("enter the weight and height please")
  a,b,c = input().split()
  a = int(a)
  b = int(b)
  c = int(c)

  for line in data:
    x = int(line[0])
    y = int(line[1])
    z = int(line[2])
    d = np.sqrt((x-a)**2+(y-b)**2+(z-c)**2)
    line.append(d)
  data.sort(key = lambda i:i[4])
  display(data)
  k = input("enter the value of k")
  k = int(k)
  sc = 0
  mc = 0
  lc = 0
  xc = 0
  for i in range(k):
    if data[i][3]=='S':
      sc+=1
    elif data[i][3]=='M':
      mc+=1
    elif data[i][3]=='L':
      lc+=1
    elif data[i][4]=='XL':
      xc+=1

  val = [sc,mc,lc,xc]
  label = ['S','M','L','XL']
  plt.bar(label,val)
  plt.xlabel('size')
  plt.ylabel('nearest neighbours')


```

```python
file = "/content/T-shirt1.csv"
temp = pd.read_csv(file)
display(temp)
fd = csv.reader(open(file))
data = []
for line in fd:
  data.append(line)


display(data)
KNN(data[1:])
```
