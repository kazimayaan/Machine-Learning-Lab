```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kmean(data):
  print("enter lightweight weight and height ")
  w1, h1 = input().split()
  print("enter middleweight weight and height ")
  w2, h2 = input().split()
  print("enter heavyweight weight and height ")
  w3, h3 = input().split()
  w1 = int(w1)
  w2 = int(w2)
  w3 = int(w3)
  h1 = int(h1)
  h2 = int(h2)
  h3 = int(h3)
  for line in data:
    x = int(line[0])
    y = int(line[1])
    d1 = np.sqrt((x-w1)**2+(y-h1)**2)
    d2 = np.sqrt((x-w2)**2+(y-h2)**2)
    d3 = np.sqrt((x-w3)**2+(y-h3)**2)

    if d1<d2 and d1<d3:
      plt.scatter(x,y,c="red")
      print(x,y," are in lightweight")
      w1 = (w1+x)/2
      h1 = (h1+y)/2
    if d2<d1 and d2<d3:
      plt.scatter(x,y,c="blue")
      print(x,y," are in middleweight")
      w2 = (w2+x)/2
      h2 = (h2+y)/2
    if d3<d2 and d3<d1:
      plt.scatter(x,y,c="green")
      print(x,y," are in heavyweight")
      w3 = (w3+x)/2
      h3 = (h3+y)/2
```

```python
file = "/content/BOXING.csv"
temp = pd.read_csv(file)
display(temp)
fd = csv.reader(open(file))
data=[]
for line in fd:
  data.append(line)
daata = data[1:]
kmean(daata)
```
