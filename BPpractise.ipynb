```python
import numpy as np
```

```python
ipn = 2
opn = 2
hln = 4
it = 6000

input = np.random.randint(1,5,ipn)
output = np.array([1.0,0.0])
hiddenlayer = np.random.rand(1,hln)

hidden_bias = np.random.rand(1,hln)
output_bias = np.random.rand(1,opn)

hidden_weight = np.random.rand(ipn,hln)
output_weight = np.random.rand(hln,opn)

def sigmoid(layer):
  return 1/(1+np.exp(-layer))

def gradient(layer):
  return layer*(1-layer)

for i in range(it):

  hiddenlayer = np.dot(input, hidden_weight)
  hiddenlayer = sigmoid(hiddenlayer+hidden_bias)

  outputlayer = np.dot(hiddenlayer, output_weight)
  outputlayer = sigmoid(outputlayer+output_bias)

  error = output - outputlayer
  gop = gradient(outputlayer)
  ghl = gradient(hiddenlayer)
  error_term_op = gop*error
  error_term_hl = ghl * np.dot(error_term_op,output_weight.T)
  
  gradient_hidden = np.dot(input.reshape(ipn,1),error_term_hl.reshape(1,hln))
  gradient_output = np.dot(hiddenlayer.reshape(hln,1),error_term_op.reshape(1,opn))

  hidden_weight = hidden_weight + 0.5*gradient_hidden
  output_weight = output_weight + 0.5*gradient_output

  print(error)
  print(outputlayer)
  print()
```
