```python
import numpy as np
```

```python
ipn = 2
opn = 2
hln = 4
iteration = 6000

input = np.random.randint(1,5,ipn)
output = np.array([1.0,0.0])
hiddenlayer = np.random.rand(1,hln)

hidden_bias = np.random.rand(1,hln)
output_bias = np.random.rand(1,opn)

hidden_weight = np.random.rand(ipn,hln)
output_weight = np.random.rand(hln,opn)

def sigmoid(layer):
  return 1/(1+np.exp(-layer))

def gradient(layer):
  return layer*(1-layer)

for i in range(iteration):
  hiddenlayer = np.dot(input,hidden_weight)
  hiddenlayer = sigmoid(hiddenlayer+hidden_bias)

  outputlayer = np.dot(hiddenlayer, output_weight)
  outputlayer = sigmoid(outputlayer+output_bias)

  error = output- outputlayer

  gop = gradient(outputlayer)
  error_term_output = gop*error

  error_term_hidden = gradient(hiddenlayer)*np.dot(error_term_output,output_weight.T)

  g_hidden = np.dot(input.reshape(ipn,1), error_term_hidden.reshape(1,hln))
  g_output = np.dot(hiddenlayer.reshape(hln,1), error_term_output.reshape(1,opn))

  hidden_weight = hidden_weight+ 0.5*g_hidden
  output_weight = output_weight + 0.5*g_output

  if i<50 or i>iteration-100:
    print("*********")
    print(error," iteration", i)
    print("#output#",outputlayer)
```
