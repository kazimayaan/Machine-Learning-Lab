```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def Bayes(DATA,x,col,yescount,nocount):
  yesc = 0
  noc = 0

  for line in DATA:
    if line[col]==x:
      if line[-1]=='yes':
        yesc+=1;
      else:
        noc+=1;
  xyes = yesc/yescount
  xno = noc/nocount

  return xyes,xno
```

```python
def main():
  file = "/content/car (2).csv"
  temp = pd.read_csv(file)
  display(temp)
  data=[]
  fd = csv.reader(open(file))
  for line in fd:
    data.append(line)
  
  DATA=data[1:]
  print("enter the car color, car type and origin:")
  x,y,z= input().split()
  yescount = 0;
  nocount=0;


  for line in DATA:
    if line[-1]=='yes':
      yescount+=1;
    else:
      nocount+=1;

  n = len(DATA[0])
  pyes=yescount/n;
  pno=nocount/n;

  xyes,xno= Bayes(DATA,x,0,yescount,nocount)
  yyes,yno= Bayes(DATA,y,1,yescount,nocount)
  zyes,zno= Bayes(DATA,z,2,yescount,nocount)

  resyes = xyes*yyes*zyes*pyes
  resno = xno*yno*zno*pno

  peryes = (resyes/(resyes+resno))*100
  perno = (resno/(resyes+resno))*100

  print("\n -----------the percentage of your car getting stolen is = ",peryes)
  print("\n -----------the percentage of your car not getting stolen is = ",perno)

  pex=[peryes,perno]
  label = ["stolen","not stolen"]

  plt.pie(pex,labels=label)
  plt.show()

```

```python
main()
```
